# some basic plots

```python
# the features - zeropadded
X = X_train[0]

label_list = ['p_t', 'p_x', 'p_y', 'p_z']

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
for i, (name, ax) in enumerate(zip(label_list, np.array(axs).ravel())):
    ax.hist(X[:,:,i].ravel(), bins=100)
    ax.set_title(name)
    ax.set_yscale("log")
```

```python
# transverse momentum of first 4 leading jets
X = X_train[0]

label_list = ['p_t_0', 'p_t_1', 'p_t_2', 'p_t_3',]

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
for i, (name, ax) in enumerate(zip(label_list, np.array(axs).ravel())):
    ax.hist(X[:,i,0].ravel(), bins=100)
    ax.set_title(name)
    ax.set_yscale("log")
```
